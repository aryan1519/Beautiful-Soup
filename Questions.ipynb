```python
import requests
from bs4 import BeautifulSoup as bs

base_url = "https://books.toscrape.com/catalogue/"

current_url = "https://books.toscrape.com/catalogue/page-1.html"

res = requests.get(current_url)

i = 0
while(i<10):
    data = bs(res.text , 'html.parser')
    next_page = data.find(class_ = 'next')
    
    next_page_url = base_url + next_page.a['href']
    
    book_name = data.find_all(class_ = 'product_pod')
    
    for j in book_name:
        print(j.h3.a['title'])
        
    current_url = next_page_url
    
    res = requests.get(current_url)
    i+=1
    
```

##### Q 13.)

```python
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs

base_url = "https://books.toscrape.com/catalogue/"

current_url = "https://books.toscrape.com/catalogue/page-1.html"

res = requests.get(current_url)

ans = []

j=0
while(j<2):
    
    data = bs(res.text , 'html.parser')
    
    next_page_url = base_url+data.find(class_ = 'next').a["href"]
    
    book_url = []
    
    l = data.find_all(class_ = 'product_pod') 
    for i in l:
        book_url.append(base_url+i.h3.a["href"])
    

    count = 0
    while(count < len(book_url)):
        
        bookURL = book_url[count]
        book_res = requests.get(bookURL)
        
        
        book_data = bs(book_res.text , 'html.parser')
        
        title = book_data.h1.string
        
        price = book_data.find(class_ = 'price_color').string
        price = float( re.search('[\d.]+' , price).group() )
        
        qty = book_data.find(class_ = 'instock availability')
        qty = qty.contents[-1].strip()
        qty = int( re.search('\d+' , qty).group() )        
        
        ans.append([title , book_url , price , qty])
        
        count+=1
    
    current_url = next_page_url
    res = requests.get(current_url)
    
    j+=1
    
    ############################
df = pd.DataFrame(ans , columns = ['Title' , 'Link' ,' Price', 'Quantity in stock' ])
df.to_csv('books.csv' , index = False)
display(df)
```
